```python
##熵权法——客观评价模型
import math                         # 导入模块
import numpy as np                  # 导入模块 numpy，并简写成 np

# 子程序：熵权法的参数设置
def ParameterSetting():
    cName = ""                  # 定义问题名称
    nVar = 2                    # 给定评价对象数量
    mVar = 3                    # 指标数量
    
    return cName, nVar, mVar



##三种指标正向化方式
def MintoMax(X,nVar,mVar):             #极小变极大
    Y = np.zeros((nVar,mVar)) 
    for j in range(mVar):
        flag = 1
        maxnum = X[0][j]
        for i in range(nVar):
            if X[i][j] > maxnum:
                maxnum = X[i][j]
            if X[i][j] < 0:
                flag = 0
        if flag == 0:
            for i in range(nVar):
                Y[i][j] = maxnum-X[i][j]
        else:
            for i in range(nVar):
                Y[i][j] = 1/X[i][j]
    return Y

def MidtoMax(X,nVar,mVar,bestMid):       #中间变极大
    Y = np.zeros((nVar,mVar)) 
    for j in range(mVar):
        maxabs = abs(X[0][j]-bestMid[j])
        for i in range(nVar):
            if maxabs < abs(X[i]-bestMid[j]):
                maxabs = abs(X[i][j]-bestMid[j])
        for i in range(nVar):
            Y[i][j] = 1 - abs(X[i][j]-bestMid[j])/maxabs
    return Y

def RantoMax(X,nVar,a,b):           #区间变极大
    Y = np.zeros((nVar,mVar))
    for j in range(mVar):
        minX = X[0][j]
        maxX = X[0][j]
        for i in range(nVar):
            if(X[i][j] > maxX):
                maxX = X[i][j]
            if(X[i][j] < minX):
                minX = X[i][j]
        M = max(a[j]-minX,maxX-b[j])
        for i in range(numVar):
            if X[i][j] < a[j] :
                Y[i][j] = 1- (a[j]-X[i][j])/M
            elif X[i][j] > b[j] :
                Y[i][j] = 1- (X[i][j]-b[j])/M
            else : Y[i][j] = 1
    return Y


##标准化处理
def Standard(X,nVar,mVar):
    Z = np.zeros((nVar,mVar)) 
    for i in range(nVar):
        maxX = X[i][0]
        minX = X[i][0]
        for j in range(mVar):
            if maxX < X[i][j]:
                maxX = X[i][j]
            if minX > X[i][j]:
                minX = X[i][j]
        for j in range(mVar):
            Z[i][j] = (X[i][j]-minX)/(maxX-minX)
    return Z

##计算概率矩阵
def Probability(Z,nVar,mVar):
    p = np.zeros((nVar,mVar)) 
    for j in range(mVar):
        sumZ = 0.0
        for i in range(nVar):
            sumZ += Z[i][j]
        for i in range(nVar):
            p[i][j] = Z[i][j]/sumZ
    return p

##计算每个指标信息熵（不确定度）
def Entropy(p,nVar,mVar):
    e = np.zeros((mVar)) 
    for j in range(mVar):
        sump = 0.0
        for i in range(nVar):
            if p[i][j] != 0:
                sump += p[i][j]*math.log(p[i][j])
        e[j] = -sump/math.log(nVar)
    return e

##计算权重
def Weight(e,mVar):
    d = np.zeros((mVar)) 
    w = np.zeros((mVar)) 
    sumd = 0.0
    for j in range(mVar):
        d[j] = 1-e[j]
        sumd += d[j]
    for j in range(mVar):
        w[j] = d[j]/sumd
    return w
        
    
##计算综合得分
def Score(w,Z,nVar,mVar):
    score = np.zeros((nVar)) 
    for i in range(nVar):
        for j in range(mVar):
            score[i] += w[j]*Z[i][j]
    return score

        
    

def main():
    [cName, nVar, mVar] = ParameterSetting()
    X = np.zeros((nVar,mVar))
    X = [[1,2,3],[3,6,1]]
    X = MintoMax(X,2,3)
    Z = Standard(X,2,3)
    p = Probability(Z,2,3)
    e = Entropy(p,2,3)
    w = Weight(e,3)
    score = Score(w,Z,2,3)
    print(score)
    
    
    
    
if __name__ == '__main__':
    main()
```
